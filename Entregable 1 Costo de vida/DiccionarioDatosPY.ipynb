```python
import wikipedia


wikipedia.set_lang("es")  

# Función para obtener la definición de una variable
def obtener_definicion_wikipedia(variable):
    try:
        
        resumen = wikipedia.summary(variable, sentences=2)
        return resumen
    
    except wikipedia.exceptions.DisambiguationError as e:
        return f"La variable '{variable}' tiene múltiples significados: {e.options}"
    except wikipedia.exceptions.PageError:
        return f"No se encontró definición para '{variable}'."
    except Exception as e:
        return f"Error: {e}"

# Conjunto de datos (lista de variables)
conjunto_datos = ["Posición (desambiguación)","País","Costo de vida","Contrato de arrendamiento","Presupuesto","Poder adquisitivo","Año"]

# Crear un diccionario para almacenar las definiciones 
definiciones = {}

# Iterar sobre el conjunto de datos y obtener la definición de cada variable
for variable in conjunto_datos:
    definiciones[variable] = obtener_definicion_wikipedia(variable)

# Mostrar las definiciones obtenidas
for variable, definicion in definiciones.items():
    print(f"Variable: {variable}, Significado: {definicion}\n")
```
